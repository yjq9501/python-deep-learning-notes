```python
# 4 创建线性张量
# arange生成的范围是[start,end)，左闭右开
# 注意：step表示步长
t8 = torch.arange(start=1,end=6,step=1)
t8
```

```python
# linspace生成的范围是[start,end]，左右都是闭区间
# 注意：steps不表示步长，表示要生成多少个元素
# t9 = torch.linspace(start=1,end=6,steps=6)
t9 = torch.linspace(start=1,end=6,steps=10)
t9
```

```python
# 5 随机张量
# 生成区间是[low,high)。左闭右开的区间
# t10 = torch.randint(low=1,high=5,size=(4,))    # 生成包含4个元素的向量张量
t10 = torch.randint(low=1,high=5,size=(2,3,4))    # 生成三维张量
t10
```

```python
# 获得随机数种子
print(torch.initial_seed())

# 手动设置随机数种子
torch.manual_seed(127)
t11 = torch.randint(low=1,high=5,size=(2,3,4))    # 生成三维张量
print(t11)
```

```python
t12 = torch.rand(4) # 随机生成指定个数的小数
t12 = torch.rand(size=(2,3,4)) # 随机生成指定形状的小数
t12
```

```python
# 随机生成小数：小数满足标准正态分布（均值为0，方差为1）
t13 = torch.randn(4)
t13 = torch.randn(size=(2,3,4))
t13
```

```python
# 6 创建指定值的张量
t14 = torch.randn(size=(2,3,4))
t14
```

```python
"""
    1- ones创建指定形状的全1张量
    2- ones_like，仿照某个张量的形状，创建指定形状的全1张量
    3- zeros创建指定形状的全0张量。【掌握】
    4- zeros_like，仿照某个张量的形状，创建指定形状的全0张量
    5- full创建指定形状，并且元素值也是人为指定的张量
    6- full_like，仿照某个张量的形状，并且元素值是人为指定的张量
"""
t15 = torch.ones(size=(3,4))    
t16 = torch.ones_like(t14)
t17 = torch.zeros(size=(3,4))
t18 = torch.zeros_like(t14)
t19 = torch.full(size=(3,4), fill_value=666)
t20 = torch.full_like(t14, fill_value=666)

print(f"ones-->{t15}")
print(f"ones_like-->{t16}")
print(f"zeros-->{t17}")
print(f"zeros_like-->{t18}")
print(f"full-->{t19}")
print(f"full_like-->{t20}")
```

```python
# 7 张量元素数据类型转换
t_1 = torch.tensor([11, 22, 33])
t_1.dtype
```

```python
# 通用的写法。掌握
t_2 = t_1.type(dtype=torch.float16) # 推荐的写法
t_2 = t_1.type(dtype=torch.DoubleTensor)
t_2
```

```python
t_3 = torch.tensor([1.678, 2.543])  # 默认是float32
print(t_3)
print(t_3.dtype)
```

```python
t_4 = t_3.half()    # float16
t_4 = t_3.short()    # 短整型int16
t_4 = t_3.int()     # int32
t_4 = t_3.long()    # 长整型int64
print(t_4.dtype)
```
