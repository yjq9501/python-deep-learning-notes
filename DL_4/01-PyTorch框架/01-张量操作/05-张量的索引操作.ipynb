```python
import torch
```

```python
t0 = torch.randint(0,10,size=(4,5))
t0
```

```python
# 【掌握】1----- 简单的行、列索引 --------
# 语法：张量对象[行索引,列索引]
# 获取第一行
print(t0[0])

# 获取第一列
print(t0[:, 0])

print(t0[2, 4])
```

```python
# 【掌握】2------ 范围索引：与容器中的切片操作完全一样 -------
# 语法：张量对象[行索引开始:行索引结束:步长, 列索引开始:列索引结束:步长]，左闭右开
# 需求1：前3行的前2列数据
print(t0[:3, :2])
```

```python
# 需求2：第2行到最后的行，并且前2列的数据
print(t0[1:, :2])
```

```python
# 【了解】3------- 列表索引 -----
# 需求1：返回(第0行,第1列)、(1,2)两个位置的元素
print(t0[[0, 1], [1, 2]])
```

```python
# 需求2：获取两个元素的坐标是(2,4)、(3,3)
# 语法：张量对象[[行1,行2], [列1,列2]]
print(t0[[2, 3], [4, 3]])
```

```python
# 获取4个元素，坐标分别是(2,4)、(2,3)、(3,4)、(3,3)
print(t0[[[2], [3]], [4, 3]])
```

```python
# 【了解】4------- 布尔索引 -------
# 需求1：第3列大于5的，对应的行数据
# 取出第3列
print(t0[:, 3])

# 判断是否是大于5
print(t0[:, 3] > 5)

# 取出对应的行数据
print(t0[t0[:, 3] > 5, :])
```

```python
# 【了解】5------ 多维索引 --------
t1 = torch.randint(0,10,size=(3,4,5))
t1
```

```python
# 获取0轴上的数据
t1[0, :, :]
```

```python
# 获取1轴上的数据
t1[:, 0, :]
```

```python
# 获取2轴上的数据
t1[:, :, 0]
```
