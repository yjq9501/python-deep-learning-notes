```python
import torch
```

```python
t1 = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float)
t1
```

```python
# sum求和。【掌握】
s_1 = t1.sum()      # 将张量中所有的元素值进行求和
s_2 = t1.sum(dim=0) # 按照第一个维度进行计算
s_3 = t1.sum(dim=1) # 按照第二个维度进行计算
# s_5 = t1.sum(dim=2) # 按照第三个维度进行计算
s_4 = t1.sum(dim=-1) # 按照最后一个维度进行计算。常用

print(f"s_1-->{s_1}")
print(f"s_2-->{s_2}")
print(f"s_3-->{s_3}")
print(f"s_4-->{s_4}")
# print(f"s_5-->{s_5}")
```

```python
# 其他计算函数
t2 = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float)
t2
```

```python
mean_result = t2.mean() # 算均值=21/6
print(mean_result)

min_result = t2.min()
print(min_result)

sqrt_result = t2.sqrt()
print(sqrt_result)

log2_result = t2.log2()
print(log2_result)
```

```python
# sum(dim)作用在三维的张量上
torch.manual_seed(129)
t9 = torch.randint(low=1,high=5,size=(2,3,4))
t9
```

```python
print("-->",t9.sum())
print("0-->",t9.sum(dim=0))
print("1-->",t9.sum(dim=1))
print("2-->",t9.sum(dim=2))
print("-1-->",t9.sum(dim=-1))
```
