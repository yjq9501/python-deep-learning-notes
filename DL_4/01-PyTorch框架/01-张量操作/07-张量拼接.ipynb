```python
import torch
```

```python
t1 = torch.randint(1,10,size=(2,3))
t2 = torch.randint(1,10,size=(2,3))
print(t1)
print(t2)
```

```python
# cat/concat
cat_1 = torch.cat(tensors=[t1,t2], dim=0)
print(cat_1)    # 4行3列

cat_2 = torch.cat(tensors=[t1,t2], dim=1)
print(cat_2)    # 2行6列

# 因为上面创建的是二维张量，因此dim的取值最大是1
# cat_3 = torch.cat(tensors=[t1,t2], dim=2)
# print(cat_3)
```

```python
# 注意：除了拼接的维度以外，其他维度必须相同
t3 = torch.randint(1,10,size=(2,3))
t4 = torch.randint(1,10,size=(4,3))   # 列数相同，因此正常
# t4 = torch.randint(1,10,size=(2,4))   # 列数不同，因此报错
print(t1)
print(t2)

cat_4 = torch.cat(tensors=[t3,t4], dim=0)
print(cat_4)
```

```python
# stack：张量的形状要完全一致
t6 = torch.randint(1,10,size=(5,6))
t7 = torch.randint(1,10,size=(5,6))
# t7 = torch.randint(1,10,size=(5,7))
# t7 = torch.randint(1,10,size=(7,6))
print(t6)
print(t7)
print(torch.stack(tensors=[t6, t7]))
```
